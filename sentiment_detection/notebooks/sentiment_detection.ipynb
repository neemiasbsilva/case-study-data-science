```python
from datasets import load_dataset
import pandas as pd
from transformers import AutoTokenizer
from transformers import AutoModel
import torch


import matplotlib.pyplot as plt
%matplotlib inline
```

# Load Dataset

```python
emotions = load_dataset("emotion")
```

```python
emotions
```

```python
emotions.set_format(type="pandas")
train_df = emotions["train"][:]
train_df.head()
```

# Exploratory Data Analysis - EDA

```python
def label_int2str(row):
    return emotions["train"].features["label"].int2str(row)

train_df["label_name"] = train_df["label"].apply(label_int2str)
train_df.head()
```

## Look distribution of classes

```python
def plot_label_distribution(dataframe):
    dataframe["label_name"].value_counts(ascending=True).plot.barh()
    plt.title("Frequency of Classes")
    plt.show()
```

```python
plot_label_distribution(train_df) # we have imbalanced dataset
```

## Duration of tweets

```python
train_df["WordsPerTweet"] = train_df["text"].str.split().apply(len)
train_df.head()
```

```python
def plot_duration(dataframe):
    dataframe.boxplot("WordsPerTweet", by="label_name", grid=False, showfliers=False,
                      color="black")
    plt.suptitle("")
    plt.xlabel("")
    plt.show()
```

```python
plot_duration(train_df)
```

# Preprocessing

```python
emotions.reset_format()
```

## Text to Tokens

```python
model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True)
```

```python
print(tokenize(emotions["train"][:3]))
```

**Look special tokens**

```python
tokens2ids = list(zip(tokenizer.all_special_tokens, tokenizer.all_special_ids))
data = sorted(tokens2ids, key=lambda x: x[-1])
special_df = pd.DataFrame(data, columns=["Special Token", "Special Token ID"])
special_df.T
```

**Encoding**

```python
emotions_encoded = emotions.map(tokenize, batched=True, batch_size=None)
```

```python
print(emotions_encoded["train"].column_names)
```
