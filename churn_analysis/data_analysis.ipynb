## Bi-variate Analysis

**Checking the relationship between CreditScore and Geography feature**

```python
fig = plt.figure(figsize=(10, 5))
sns.boxplot(x="CreditScore", y="Geography", data=df)
plt.show()
```

**Checking the relationship between CreditScore and Gender feature**

```python
fig = plt.figure(figsize=(10, 5))
sns.boxplot(x="CreditScore", y="Gender", data=df)
plt.show()
```

**Checking relationship between CreditScore and Geography based on gender and size using the salary feature**

```python
fig = plt.figure(figsize=(10,5))
sns.scatterplot( x="CreditScore", y='Geography', data=df,
                hue='Gender', size='EstimatedSalary')
  
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
  
plt.show()
```
